```python
output_path = "/mnt/c/Users/mavi/AMC/Results/"
import json
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
data= open(output_path+'23_signals/length_barker.json')
length_barker = json.load(data)
```

```python
data= open(output_path+'23_signals/length_costas.json')
length_costas = json.load(data)
```

```python
len(length_barker)
```

```python
import numpy as np
#plt.hist(length_costas)S
barWidth = 0.6
codes = [3,4,5,6,7,11,13]
bins = [x+n for n in range(3, 13) for x in [0.0, 0.25]]+[13]
bins_2 = [x+n for n in range(3, 13) for x in [0.25, 0.5]]+[13]
plt.hist(length_barker, weights = np.ones(len(length_barker))/len(length_barker), alpha=0.6, width=barWidth)
plt.hist(length_costas,weights = np.ones(len(length_costas))/len(length_costas), alpha=0.6, width=barWidth)
plt.xticks(codes, fontsize=10.5)
plt.yticks(np.arange(0, 0.6, 0.05), fontsize=10.5)
#plt.xticks([r + barWidth for r in range(len(codes))],["3", "4", "5", "6", "7", "11", "13"])
plt.grid(color = "grey", linewidth = "0.8", linestyle = "--", axis='y')
plt.legend(["Barker classified as Costas", "Costas classified as Barker"], fontsize=11.5)
plt.xlabel("Number of symbols", fontsize=11.5)
plt.ylabel("% of the misclassified signals", fontsize=11.5)
#plt.xticks([5,7,11,13])
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.savefig(output_path + '23_signals/number_of_symbols.png')
```

```python
barWidth=1
rates = [1, 2, 5, 10, 15, 20]
bins = [x+n for n in range(1, 20) for x in [0.0, 0.25]]+[20]
bins_2 = [x+n for n in range(1, 20) for x in [0.25, 0.5]]+[20]
plt.hist(vs_barker,bins_2,weights = np.ones(len(vs_barker))/len(vs_barker), alpha=0.6, width=barWidth)
plt.hist(vs_costas, bins,weights = np.ones(len(vs_costas))/len(vs_costas), alpha=0.6, width=barWidth)
plt.xticks(rates, fontsize=10.5)
plt.yticks(np.arange(0, 0.6, 0.05), fontsize=10.5)
#plt.xticks([r + barWidth for r in range(len(codes))],["3", "4", "5", "6", "7", "11", "13"])
plt.grid(color = "grey", linewidth = "0.8", linestyle = "--", axis='y')
plt.legend(["Barker classified as Costas", "Costas classified as Barker"], fontsize=11.5)
plt.xlabel("Symbol rate (Msymb/s)", fontsize=11.5)
plt.ylabel("% of the misclassified signals", fontsize=11.5)
#plt.xticks([5,7,11,13])
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))
plt.savefig(output_path + '23_signals/symbol_rate.png')
```
